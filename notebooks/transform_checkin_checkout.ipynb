# Cleaning plan of checkin_checkout

```python
import pandas as pd
```

### Nulls?

```python
checkin_checkout = pd.read_csv("../data/raw/checkin_checkout_history.csv")

checkin_checkout.info()
```

```python
# Check for nulls
checkin_checkout = checkin_checkout.replace(" ", None)
print("Num of nulls: " + str(checkin_checkout.isnull().any(axis=1).sum()))
print(checkin_checkout.shape)
```

```python
print("Number of missing values per column")
for col in checkin_checkout.columns:
    print(f"{col}: {checkin_checkout[col].isnull().sum()}")
```

> 1269 ~ 0.5% (0.4229...%) -> Not a lot thus it can be removed

```python
checkin_checkout = checkin_checkout.dropna()
print("Num of nulls: " + str(checkin_checkout.isnull().any(axis=1).sum()))
print(checkin_checkout.shape)
checkin_checkout.info()
```

### Correct data types?

```python
checkin_checkout = pd.read_csv("../data/raw/checkin_checkout_history.csv")

checkin_checkout = checkin_checkout.replace(" ", None)
checkin_checkout = checkin_checkout.dropna()
checkin_checkout.info()
```

```python
checkin_checkout.head()
```

```python
checkin_checkout["checkin_time"] = checkin_checkout["checkin_time"].astype(
    "datetime64[ns]"
)
checkin_checkout["checkout_time"] = checkin_checkout["checkout_time"].astype(
    "datetime64[ns]"
)
print(checkin_checkout.info())
checkin_checkout.head()
```

```python
# Possibly a better way...
checkin_checkout = checkin_checkout.astype(
    {"checkin_time": "datetime64[ns]", 
     "checkout_time": "datetime64[ns]"}
)

print(checkin_checkout.info())
checkin_checkout.head()
```

```python
temp_list = []
for x in checkin_checkout["calories_burned"]:
    temp_list.append(x.is_integer())
series = pd.Series(temp_list)
print(series.sum())
print(checkin_checkout.shape)
```

> Thus all the values are integers

```python
checkin_checkout["calories_burned"] = checkin_checkout[
    "calories_burned"
].astype(int)

print(checkin_checkout.info())
checkin_checkout.head()
```

### Duplicates?

```python
checkin_checkout = pd.read_csv("../data/raw/checkin_checkout_history.csv")

checkin_checkout = checkin_checkout.replace(" ", None)
checkin_checkout = checkin_checkout.dropna()
checkin_checkout = checkin_checkout.astype(
    {"checkin_time": "datetime64[ns]", "checkout_time": "datetime64[ns]", "calories_burned": int}
)
checkin_checkout.info()
```

```python
print("Num of duplicates: " + str(checkin_checkout.duplicated().sum()))
print(checkin_checkout[checkin_checkout.duplicated()])
```

> 57 ~ 0.019% -> Can be removed

```python
checkin_checkout = checkin_checkout.drop_duplicates()
print("Num of duplicates: " + str(checkin_checkout.duplicated().sum()))
print(checkin_checkout.shape)
```

### Duration column

```python
checkin_checkout = pd.read_csv("../data/raw/checkin_checkout_history.csv")

checkin_checkout = checkin_checkout.replace(" ", None)
checkin_checkout = checkin_checkout.dropna()
checkin_checkout = checkin_checkout.astype(
    {
        "checkin_time": "datetime64[ns]",
        "checkout_time": "datetime64[ns]",
        "calories_burned": int,
    }
)
checkin_checkout = checkin_checkout.drop_duplicates()
checkin_checkout["user_id"] = checkin_checkout["user_id"].str.strip()
checkin_checkout.info()
```

```python
checkin_checkout["duration"] = (
    checkin_checkout["checkout_time"] - checkin_checkout["checkin_time"]
)
checkin_checkout.info()
checkin_checkout.head()
```
