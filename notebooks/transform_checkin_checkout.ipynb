# Exploratory Data analysis of checkin_checkout

```python
import pandas as pd
```

### Nulls?

#### Checkin_checkout

```python
checkin_checkout = pd.read_csv("../data/raw/checkin_checkout_history.csv")

checkin_checkout.info()
```

```python
# Check for nulls
print(checkin_checkout.isnull().any(axis=1).sum())
print(checkin_checkout.shape)
```

```python
print("Number of missing values per column")
for col in checkin_checkout.columns:
    print(f"{col}: {checkin_checkout[col].isnull().sum()}")
```

> 1269 ~ 0.5% (0.4229...%) -> Not a lot thus it can be removed

```python
checkin_checkout = checkin_checkout.dropna()
print(checkin_checkout.isnull().any(axis=1).sum())
print(checkin_checkout.shape)
```
