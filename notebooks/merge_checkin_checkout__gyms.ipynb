# Merge plan of checkin_checkout__gyms

```python
import pandas as pd
```

```python
checkin_checkout = pd.read_csv("../data/raw/checkin_checkout_history.csv")

checkin_checkout = checkin_checkout.replace(" ", None)
checkin_checkout = checkin_checkout.dropna()
checkin_checkout = checkin_checkout.astype(
    {
        "checkin_time": "datetime64[ns]",
        "checkout_time": "datetime64[ns]",
        "calories_burned": int,
    }
)
checkin_checkout = checkin_checkout.drop_duplicates()
checkin_checkout["user_id"] = checkin_checkout["user_id"].str.strip()
checkin_checkout["duration"] = (
    checkin_checkout["checkout_time"] - checkin_checkout["checkin_time"]
)
checkin_checkout.info()
```

```python
gyms = pd.read_csv("../data/raw/gym_locations_data.csv")

gyms = gyms.drop("facilities", axis=1)
gyms.info()
```

```python
merged_data = pd.merge(checkin_checkout, gyms, on="gym_id")
merged_data.info()
```
