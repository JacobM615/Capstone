# Merge plan of users__sub_plans

```python
import pandas as pd
```

```python
users = pd.read_csv("../data/raw/users_data.csv")

users = users.replace(" ", None)
users = users.dropna()
users = users.astype(
    {
        "age": int,
        "sign_up_date": "datetime64[ns]",
    }
)
users = users.drop_duplicates()
users = users.drop("birthdate", axis=1)
users.info()
```

```python
sub_plans = pd.read_csv("../data/raw/subscription_plans.csv")

sub_plans = sub_plans.drop("features", axis=1)
sub_plans.info()
```

```python
merged_data = pd.merge(users, sub_plans, on="subscription_plan")
merged_data.info()
```

```python
merged_data.to_csv(
    "../tests/test_data/expected/users__sub_plans.csv", index=False
)
```
